            background-color: white;

            font-family: sans-serif;

            body {

                  }

Question 7:  Explain the difference between position: absolute; and position: fixed;
in CSS.

- **position: absolute ➡**
        An absolutely positioned element is placed relative to its nearest "positioned" ancestor. A positioned ancestor is any parent, grandparent, etc., that has its position property set to anything other than the default (static), such as relative, absolute, fixed, or sticky.
        
      


**Scroll Behavior** ➡ The element scrolls along with the rest of the page content. If its positioned ancestor scrolls out of view, the absolutely positioned element will also scroll out of view.



**Document Flow** ➡ It is completely removed from the normal document flow. This means it no longer takes up space, and other elements will render as if it doesn't exist, which can cause content to overlap.


- **position: fixed  ➡**
Reference Point: A fixed-position element is placed relative to the browser's viewport (the visible area of the window). This means its position is independent of any parent elements.


**Scroll Behavior ➡** It stays in the same position on the screen even when the page is scrolled. This is what creates a "floating" effect.
Document Flow: Like position: absolute;, it is removed from the normal document flow and will not affect the layout of surrounding elements.
Use Case: Ideal for persistent UI elements that should always be visible, such as a sticky navigation bar, a "back to top" button, or a chat widget.
